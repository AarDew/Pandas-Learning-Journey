# Merge Join Concatenate

```python
import pandas as pd
```

```python
df1=pd.read_csv(r"D:\LOTR.csv")
df1
```

```python
df2=pd.read_csv(r"D:\LOTR 2.csv")
df2
```

```python
df1.merge(df2)  # inner join
```

```python
df1.merge(df2,how='inner')  # same as above
```

```python
df1.merge(df2,how='inner',on='FellowshipID')
```

```python
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how='outer')   # outer join
```

```python
df1.merge(df2,how='left')   #left join 
```

```python
df1.merge(df2,how='right')   #right join 
```

```python
df1.merge(df2,how='cross')   #cros join 
```

```python
df1.join(df2)   # join is similar to merge but its a little bit difficult to implement
```

```python
df1.join(df2,on='FellowshipID',how='outer',lsuffix='_Left',rsuffix='_Right')   #its better while working with index
```

```python
df4=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left',rsuffix='_Right')
df4
```

```python
df4=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left',rsuffix='_Right',how='outer')
df4
```

```python
# now concat
# what it does is it puts data of one df top of another df unlike merge and join which puts side by side
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2],join='inner',axis=1)
```
