# Group by and Aggregating

```python
import pandas as pd
```

```python
df=pd.read_csv(r"D:\Flavors.csv")
```

```python
df
```

```python
df.groupby('Base Flavor')
```

```python
group_by_frame=df.groupby('Base Flavor')
```

```python
group_by_frame.mean()
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
 df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
 df.groupby('Base Flavor').count()
```

```python
 df.groupby('Base Flavor').min()
```

```python
 df.groupby('Base Flavor').max()
```

```python
 df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
 df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum'],'Texture Rating':['mean','max','count','sum']})
```

```python
df
```

```python
# now group by on multiple columns
```

```python
df.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum'],'Texture Rating':['mean','max','count','sum']})
```
